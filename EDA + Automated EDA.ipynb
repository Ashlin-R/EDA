- EDA
    - pd.pandas.set_option('display.max_columns',None) [SHOW ALL COLUMNS]

	- value_counts()
	- isnull()

	- duplicated() , drop_duplicates()
	- Correlation (sns.heatmap(df.corr()))
	- NUMERICAL FEATURES:
        - find distribution of numerical features, if skewed then transform
		- jointplot (relationship between two numerical features) - scatter/hex/reg
		- pairplot (relationship between multiple features) 
		- distplot (distribution of numerical feature) - kde=True/False, Percentage/Count - use to count as well (bins)
		- hue = split based on a categorical feature (eg: sex)
	- CATEGORICAL FEATURES:
        - cardinality
		- countplot 
		- barplot (relationship between categorical(x)[sex], numeric(y)[total_bill], hue(smoker)) 
		- boxplot (min,25,median,75,max,outliers of categorical(x)[sex] wrt numeric(y)[total_bill], hue(smoker))
		- violinplot (boxplot + distribution(KDE) of the data : x=total_bill , y=sex)

```python
train.isnull().sum()
```

```python
null_list = train.isnull().sum()
for a,b in enumerate(null_list.iteritems()):
    print(f"{b[0]} - {(b[1]/len(train))*100}%")
```

```python
n_list = []
for a,b in enumerate(null_list.iteritems()):
    if b[1]!=0:
        n_list.append(b)

for a,b in n_list:
    #print(f"{a} - {(b/len(train))*100}%")
    #null_df = pd.DataFrame(data={'Feature': [b[0]], 'Percentage of null values': [(b[1]/len(train))*100]})
    print("{feature} - {percentage:1.2f}%".format(feature=a,percentage=(b/len(train))*100))
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived', hue='Sex', data=train)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived', hue='Pclass', data=train)
```

```python
sns.distplot(train['Age'].dropna(), kde=False, bins=10)
```

```python
sns.boxplot(x='Pclass',y='Age',data=train)
```

```python
train.groupby(train['Pclass']).mean()
```

```python
#Need to obtain mean algorithimcally
def impute_age(cols):
    
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        elif Pclass==3:
            return 24
    else:
        return Age    
```

```python
train['Age'] = train[['Age' , 'Pclass']].apply(impute_age, axis=1)
```

```python
train.isnull().sum()
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
train.drop('Cabin', axis=1, inplace=True)
```

```python
train.head()
```

```python
pd.get_dummies(train['Embarked'], drop_first=True).head()
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
embark = pd.get_dummies(train['Embarked'], drop_first=True)
```

```python
train.drop(['Sex','Embarked','Name','Ticket'], axis=1, inplace=True) 
```

```python
train.head()
```

```python
train = pd.concat([train,embark,sex],axis=1) 
```

```python
train.head()
```

```python
X = train.drop('Survived',axis=1)
```
